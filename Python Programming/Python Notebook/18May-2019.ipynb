#### Problem 3:
#### Problem statement:
Line graph for income of California

```python
import pandas as pd
df = pd.read_csv('DataFiles/Income.csv')
y = df.values[4]
x = df.columns
#x
#y
```

```python
import matplotlib.pyplot as plt
years = []
for i in range(2, len(x)):
    years.append(x[i])
years
Income = []
for i in range(2, len(y)):
    Income.append(y[i])
Income

plt.xlabel('years')
plt.ylabel('Income')
plt.title('Income of california state')
plt.plot(years, Income)
plt.savefig('Figures/line-graph-Income-california.png')
plt.show()
```

#### Problem 4:
#### Problem statement:
Generate Scatter plot for 50 different numbers

```python
import numpy as np
n = 50
x = np.random.rand(n) # Generate 50 random numbers
y = np.random.rand(n) # Generate 50 random numbers
colorlist = np.random.rand(n) # GEnerate 50 random colors
arealist = np.pi * (15 * np.random.rand(50))**2 # Generate 50 random areas of circles
plt.scatter(x, y, s = arealist, c = colorlist, alpha = 1)
plt.title('50 Random Numbers')
plt.savefig('figures/50-random-numbers.png')
plt.show()
```

#### Problem 5:
#### Problem statement:
Piechart using matplotlib

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
